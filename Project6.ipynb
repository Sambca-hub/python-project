6. Quiz Engine with Dynamic Scoring
Create a Quiz Engine that dynamically evaluates participants.
The system should manage a question bank categorized by difficulty levels. During quiz
attempts, questions should appear in a pseudo-random order without using random
libraries.
Scoring must vary by difficulty, include negative marking, and generate detailed score
reports. String handling and nested data structures must be used extensively

```python
class QuizEngine:
    def __init__(self, question_bank):

        self.question_bank = question_bank
        self.user_answers = []
        self.score = 0

        self.results_by_difficulty = {
            'easy': {'correct': 0, 'incorrect': 0, 'score': 0},
            'medium': {'correct': 0, 'incorrect': 0, 'score': 0},
            'hard': {'correct': 0, 'incorrect': 0, 'score': 0}
        }

        self.detailed_results = []

    def _gcd(self, a, b):

        while b:
            a, b = b, a % b
        return a

    def _generate_question_order(self):

        num_questions = len(self.question_bank)
        if num_questions == 0:
            return []
        multiplier = 3
        while self._gcd(multiplier, num_questions) != 1:
            multiplier += 1
            if multiplier > num_questions * 2 + 1:
                multiplier = 1
                break

        pseudo_random_order = []
        for i in range(num_questions):
            pseudo_random_order.append((i * multiplier) % num_questions)
        return pseudo_random_order

    def _present_question(self, question_data):

        question_text = question_data["question"]
        options = question_data["options"]
        correct_answer = question_data["answer"]
        difficulty = question_data["difficulty"]
        positive_score = question_data["positive_score"]
        negative_score = question_data["negative_score"]

        print(f"\nDifficulty: {difficulty.capitalize()}")
        print(f"Question: {question_text}")

        for i, option in enumerate(options):
            print(f"{i+1}. {option}")


        user_input = input("Your answer (enter option number): ").strip()
        user_answer_text = "N/A (Invalid Input)"
        is_correct = False
        current_q_score = 0

        try:
            choice_index = int(user_input) - 1
            if 0 <= choice_index < len(options):
                user_answer_text = options[choice_index]

                if user_answer_text.lower() == correct_answer.lower():
                    is_correct = True
            else:
                print("Invalid option number. Considering as incorrect.")
        except ValueError:
            print("Invalid input. Please enter a number. Considering as incorrect.")

        if is_correct:
            current_q_score = positive_score
            self.score += positive_score
            self.results_by_difficulty[difficulty]['correct'] += 1
            self.results_by_difficulty[difficulty]['score'] += positive_score
            print(f"Correct! You gained {positive_score} points.")
        else:
            current_q_score = negative_score
            self.score += negative_score
            self.results_by_difficulty[difficulty]['incorrect'] += 1
            self.results_by_difficulty[difficulty]['score'] += negative_score

            print(f"Incorrect. You lost {abs(negative_score)} points. The correct answer was '{correct_answer}'.")


        self.detailed_results.append({
            "question": question_text,
            "user_answer": user_answer_text,
            "correct_answer": correct_answer,
            "is_correct": is_correct,
            "score_change": current_q_score,
            "difficulty": difficulty
        })

    def start_quiz(self):

        question_order = self._generate_question_order()
        print("--- Quiz Started ---")
        if not question_order:
            print("No questions in the bank to start the quiz.")
            return

        for q_index in question_order:
            question_data = self.question_bank[q_index]
            self._present_question(question_data)

        print("\n--- Quiz Finished ---")
        self.generate_report()

    def generate_report(self):

        print("\n--- Quiz Report ---")
        print(f"Total Score: {self.score}")

        print("\nBreakdown by Difficulty:")

        for difficulty, stats in self.results_by_difficulty.items():
            print(f"  {difficulty.capitalize()}:")
            print(f"    Correct: {stats['correct']}")
            print(f"    Incorrect: {stats['incorrect']}")
            print(f"    Score: {stats['score']}")

        print("\nDetailed Question Results:")
        if not self.detailed_results:
            print("  No questions were attempted.")
        for i, result in enumerate(self.detailed_results):
            status = "Correct" if result['is_correct'] else "Incorrect"

            print(f"\nQuestion {i+1}: {result['question']}")
            print(f"  Your Answer: '{result['user_answer']}'")
            print(f"  Correct Answer: '{result['correct_answer']}'")
            print(f"  Status: {status} ({result['score_change']} points)")
            print(f"  Difficulty: {result['difficulty'].capitalize()}")

        print("\n--- End of Report ---")


question_bank_data = [
    {
        "question": "What is the capital of France?",
        "options": ["Berlin", "Madrid", "Paris", "Rome"],
        "answer": "Paris",
        "difficulty": "easy",
        "positive_score": 10,
        "negative_score": -3
    },
    {
        "question": "Which planet is known as the Red Planet?",
        "options": ["Earth", "Mars", "Jupiter", "Venus"],
        "answer": "Mars",
        "difficulty": "easy",
        "positive_score": 10,
        "negative_score": -3
    },
    {
        "question": "What is the largest ocean on Earth?",
        "options": ["Atlantic", "Indian", "Arctic", "Pacific"],
        "answer": "Pacific",
        "difficulty": "medium",
        "positive_score": 20,
        "negative_score": -7
    },
    {
        "question": "Who wrote 'Romeo and Juliet'?",
        "options": ["Charles Dickens", "William Shakespeare", "Jane Austen", "Mark Twain"],
        "answer": "William Shakespeare",
        "difficulty": "medium",
        "positive_score": 20,
        "negative_score": -7
    },
    {
        "question": "What is the chemical symbol for gold?",
        "options": ["Au", "Ag", "Fe", "Hg"],
        "answer": "Au",
        "difficulty": "hard",
        "positive_score": 30,
        "negative_score": -10
    },
    {
        "question": "In which year did the first human land on the Moon?",
        "options": ["1965", "1969", "1971", "1975"],
        "answer": "1969",
        "difficulty": "hard",
        "positive_score": 30,
        "negative_score": -10
    }
]

my_quiz = QuizEngine(question_bank_data)

print("Welcome to the Quiz!")
my_quiz.start_quiz()
```
